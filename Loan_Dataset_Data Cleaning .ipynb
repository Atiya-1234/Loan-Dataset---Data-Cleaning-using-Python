## Detecting Outlier using IQR(Inter Quantile Range)

```python
df.describe()
```

```python
df.shape
```

```python
q1=df['CoapplicantIncome'].quantile(0.25)
q1
```

```python
q3=df['CoapplicantIncome'].quantile(0.75)
q3
```

```python
IQR=q3-q1
IQR
```

```python
min_range=q1-(1.5*IQR)
min_range
```

```python
max_range=q3+(1.5*IQR)
max_range
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='CoapplicantIncome',data=df)
plt.show()
```

## Removing Outlier

```python
new_df=df[df['CoapplicantIncome']<=max_range]
```

```python
new_df
```

```python
new_df.shape
```

```python
sns.boxplot(x='CoapplicantIncome',data=new_df)
plt.show()
```

## Function Transformer

```python
df=pd.read_csv("D:\\Python_Notes\\loan.csv")
df.head()
```

```python
df.info()
```

```python
sns.distplot(df['CoapplicantIncome'])
```

```python
q1=df['CoapplicantIncome'].quantile(0.25)
q3=df['CoapplicantIncome'].quantile(0.75)
iqr=q3-q1
```

```python
min_r=q1-(1.5*iqr)
max_r=q3+(1.5*iqr)
min_r,max_r
```

```python
df=df[df['CoapplicantIncome']<=max_r]
```

```python
sns.distplot(df['CoapplicantIncome'])
plt.show()
```

```python
## Function Transformation
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer(func=np.log1p)
ft.fit(df[['CoapplicantIncome']])
```

```python
df['CoapplicantIncome_ft']=ft.transform(df[['CoapplicantIncome']])
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Before')
sns.distplot(df['CoapplicantIncome'])
plt.subplot(1,2,2)
plt.title('After')
sns.distplot(df['CoapplicantIncome_ft'])
plt.show()
```
